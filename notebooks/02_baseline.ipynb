# Baseline Model — TF-IDF + Logistic Regression

```python
import sys, json
sys.path.insert(0, '..')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yaml

from src.data.dataset import load_splits
from src.models.baseline import build_pipeline, train, evaluate

with open('../config/config.yaml') as f:
    cfg = yaml.safe_load(f)

train_df, val_df, test_df = load_splits('../data/processed')
```

## Train Baseline

```python
pipeline = train(train_df, val_df, cfg, save_dir='../models/baseline')
report = evaluate(pipeline, test_df, results_dir='../results')
```

## Results Summary

```python
weighted_f1 = report['weighted avg']['f1-score']
accuracy = report['accuracy']
print(f'Weighted F1:  {weighted_f1:.4f}')
print(f'Accuracy:     {accuracy:.4f}')
print()
print('Per-class F1:')
for label in sorted(report.keys()):
    if label not in ('accuracy', 'macro avg', 'weighted avg'):
        print(f'  {label:<25} {report[label]["f1-score"]:.4f}')
```
